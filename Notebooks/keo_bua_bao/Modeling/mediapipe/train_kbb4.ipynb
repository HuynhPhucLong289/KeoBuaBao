```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv("data\\keo_bua_bao\\raw\\toa_do_kbb.csv", header = None)         #Truy cập file 
df = df.dropna()                                                          #Loại bỏ giá trị thiếu
```

```python
df.head()       #5 dòng đầu của file tạo độ
```

```python
x_raw = df.iloc[:, 1:].values                                             #Lấy tất cả giá trị ở cột 1 tới cuối
y_raw = df.iloc[:, 0].values                                              #Lấy tất cả giá trị ở cột 0
```

```python
def khoang_cach_toa_do(toa_do_tay):
    toa_do = toa_do_tay.reshape(21,3)                           #chuyen tu mang 1 chieu sang 2 chieu
    co_tay = toa_do[0]                                          #lay x,y,z co tay
    chieu_dai = np.linalg.norm(toa_do[9] - co_tay) + 1e-6       #tinh chieu dai ban tay (dau ngon giua - co tay)

    chuan_hoa = (toa_do - co_tay)/chieu_dai                     #dua ban tay ve vi tri (0,0)

    ngon_tay = [4,8,12,16,20]                                   #tọa độ các đầu ngón tay
    khoang_cach = []
    for i in ngon_tay:
        khoang_cach_co_tay = np.linalg.norm((toa_do[i] - co_tay)/chieu_dai) #tinh khoảng các từ các đầu ngón tay tới cổ tay (dùng công thức euclid)
        khoang_cach.append(khoang_cach_co_tay)         
    for i in range(len(ngon_tay)-1):                
        khoang_cach_ngon_tay = np.linalg.norm(toa_do[ngon_tay[i]] - toa_do[ngon_tay[i+1]])  #tính khoảng cách giữa các đầu ngón tay
        khoang_cach.append(khoang_cach_ngon_tay)
    
    return np.hstack([chuan_hoa.flatten(), khoang_cach])        #ghép các tọa độ đã được chuẩn hóa với khoảng cách giữa các ngón tay, cổ tay 
```

```python
m = []                                                     #chuẩn hóa tọa độ của tất cả các điểm
for i in x_raw:
    x_chuan_hoa = m.append(khoang_cach_toa_do(i)) 
x_chuan_hoa1 = np.array(m)
```

```python
n = np.unique(y_raw)                        #chuyen y_raw thanh numpy
```

```python
chuyen_thanh_so = {}                        #chuyen thanh dict {a:1, b:0}
for i,lable in enumerate(n):
    chuyen_thanh_so[lable] = i
```

```python
y_list = []                                 #Biến mảng ['a', 'a', 'b'] thành [0, 0, 1]
for key in y_raw:
    value = chuyen_thanh_so[key]
    y_list.append(value)
y_int = np.array(y_list)  
```

```python
cot = len(n)
onehot = np.zeros((len(y_int), cot))        #tao 1 mang toan 0 voi y_int hang va y_raw cot)
onehot[np.arange(len(y_int)), y_int] = 1    #gan gia tri o hang len(y_int) cot y_int thanh 1
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_chuan_hoa1, onehot, test_size=0.2, random_state=42) #chia file train, test
```

```python
model = keras.Sequential([                      #tạo khung cho bộ não model
    layers.Input(shape = (72,)),                #tạo input
    layers.Dense(256, activation = 'relu'),     #tạo lớp dense lấy đặc điểm
    layers.Dropout(0.2),                        #tắt ngẫu nhiên 20% no ron ở lớp trên
    layers.Dense(128, activation = 'relu'),
    layers.BatchNormalization(),               
    layers.Dense(64, activation = 'relu'),      #tổng các các đặc điểm từ lớp dense ở trên
    layers.Dense(len(n), activation = 'softmax')     #tong hop cac lop o tren roi chuyen ket qua ve dang%
])             
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics = ['accuracy'])          
history = model.fit(x_train, y_train, epochs = 15, batch_size = 32, validation_data = (x_test, y_test))
```
